You are given a list of temperatures (in °C) recorded over a week:

temperatures = [22, 29, 31, 18, 35, 25, 19]


Write a Python program to do the following:

Separate the temperatures into two lists:

hot_days: for temperatures 30°C or above

normal_days: for temperatures below 30°C

Create a class named WeatherStats with the following methods:

__init__(self, temps): Initializes the object with a list of temperatures.

highest(self): Returns the highest temperature.

lowest(self): Returns the lowest temperature.

average(self): Returns the average temperature.

display(self): Displays the temperature list, highest, lowest, and average temperature.

Create two objects of the WeatherStats class:

One for hot_days

One for normal_days

Use the display() method to show statistics for both hot and normal days.

```python
n = int(input("Temperature number? "))

temp = []

for i in range(n):
    value = float(input(f"Enter temperature #{i+1}: "))
    temp.append(value)

hot_day = list()
normal_day = list()

for i in temp:
    if i >= 30:
        hot_day.append(i)
    else:
        normal_day.append(i)

class WeatherStats:
    def __init__(self, temps):
        self.temps = temps

    def highest(self):
      return max(self.temps)

    def lowest(self):
      return min(self.temps)

    def average(self):
      return sum(self.temps)/len(self.temps)

    def display(self):
      print(f"Temperatures: {self.temps}")
      print("Highest Temperatures:", self.highest())
      print("Lowest Temperatures:", self.lowest())
      print("Average Temperature:", self.average())

hot_temp_stat = WeatherStats(hot_day)
normal_temp_stat = WeatherStats(normal_day)

hot_temp_stat.display()
normal_temp_stat.display()
```
